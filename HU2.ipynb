```python
import pandas as pd
import numpy as np  
```

```python
df = pd.read_csv("datos/columnas_def.csv")
df.head()
```

```python
def whitespace_remover(df):
   
    # iterating over the columns
    try:
        
        for i in df.columns:
         
        # checking datatype of each columns
            if df[i].dtype == 'object':
             
            # applying strip function on column
                df[i] = df[i].map(str.strip)
        else: 
            pass
                
    except:
        pass
# applying whitespace_remover function on dataframe
whitespace_remover(df)
 
# printing dataframe
df.head(2)
```

```python
df = df.replace(np.nan,"null")

columnas = ["Q10","Q18", "Q19","Q27","Q29", "Q30", "Q31", "Q32", "Q34", "Q36", "Q37", "Q38","Q39", "Q40", "Q42"]

for values in columnas:
    columnas_juntar = [x for x in list(df.columns.values) if values in x]
    df[values] = df[columnas_juntar].agg(",".join, axis = 1)
    df = df.drop(columns = columnas_juntar)
```

```python
patron = "null(,|s|.*)"

df.replace(to_replace=patron, value='', regex=True,inplace=True)
```
