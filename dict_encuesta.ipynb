```python
# defino dos listas, de questions y de enunciados de cada pregunta. Lo hago estableciendo patrones de regex para ir separando las partes que me interesan del texto. 
# guardo la búsqueda de regex en una variable (en este caso x, y) para poder apendearla a la lista después. 

patron_q = "(Q.*)\n"
patron_e = "-[B-Z].*\n|-[B-Z].*\?"
```

```python
# lista de enunciados
y = re.findall(patron_e, encuesta)
```
