```python
# me genero un diccionario con la unión de las columnas por partes, en las Q_finales. Los valores los incluyo en una lista.
#  voy a probar con un regex
"""
FUNCIONA
"""

for i in lista_Q:
    lista2 = []
    patron = f"({i})"
    for j in df_k2.columns:
        
        if re.search(patron,j):
            for x in df_k2[j].values:
                if x == "False":
                    pass
                else:
                    lista2.append(x)
                    dict_2[i]=lista2
```
